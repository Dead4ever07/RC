```python
import matplotlib.pyplot as plt

FILE_SIZE = 87744
L = [256,512,1024,2048]
BER = [0,2.5e-5, 5e-5,7.5e-5]
DEL = [0,0.01,0.049999,0.099999]
BAUD = [1800, 9600,38400,115200]

DEF_L = 8192
DEF_BER = 0
DEF_DEL = 0
DEF_BAUD = 9600

Se_l = [74.67,76.78,77.85,78.32]
Se_ber = [77.85,65.79,57.63,53.36]
Se_del = [77.85,75.81,69.03,45.34]
Se_baud = [77.84,77.85,77.82,77.80]


def S(l, ber,c,Tprop):
    n = (1-ber)**l
    tf = l/c
    a = Tprop/tf
    d = 1 + 2*a
    return n/d*100


###################################################
# 1) Vary L
###################################################
theoretical_L = [S(l, DEF_BER, DEF_BAUD, DEF_DEL) for l in L]

plt.figure()
plt.plot(L, theoretical_L, label="Teórica")
plt.plot(L, Se_l, label="Experimental", linestyle="--")
plt.xlabel("L (bytes)")
plt.ylabel("Eficiência (%)")
plt.title("Eficiência vs L")
plt.legend()
plt.grid(True)
plt.ylim(0, 100)
plt.show()

###################################################
# 2) Vary BER
###################################################
theoretical_BER = [S(DEF_L, ber, DEF_BAUD, DEF_DEL) for ber in BER]

plt.figure()
plt.plot(BER, theoretical_BER, label="Teórica")
plt.plot(BER, Se_ber, label="Experimental", linestyle="--")
plt.xlabel("BER")
plt.ylabel("Eficiência (%)")
plt.title("Eficiência vs BER")
plt.legend()
plt.grid(True)
plt.ylim(0, 100)
plt.show()

###################################################
# 3) Vary Propagation Delay (Tprop)
###################################################
theoretical_DEL = [S(DEF_L, DEF_BER, DEF_BAUD, d) for d in DEL]

plt.figure()
plt.plot(DEL, theoretical_DEL, label="Teórica")
plt.plot(DEL, Se_del, label="Experimental", linestyle="--")
plt.xlabel("Tprop (s)")
plt.ylabel("Eficiência (%)")
plt.title("Eficiência vs Tprop")
plt.legend()
plt.grid(True)
plt.ylim(0, 100)
plt.show()

###################################################
# 4) Vary Baud Rate (c)
###################################################
theoretical_BAUD = [S(DEF_L, DEF_BER, b, DEF_DEL) for b in BAUD]

plt.figure()
plt.plot(BAUD, theoretical_BAUD, label="Teórica")
plt.plot(BAUD, Se_baud, label="Experimental", linestyle="--")
plt.xlabel("Baud (bps)")
plt.ylabel("Eficiência (%)")
plt.title("Eficiência vs Baud Rate")
plt.legend()
plt.grid(True)
plt.ylim(0, 100)
plt.show()
```
