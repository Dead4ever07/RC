```python
FILE_SIZE = 87744

L = [2048,4096,8192,16384]
S_l = [74.67,76.78,77.85,78.32]
t_l = [11904, 12240,11742 , 11670]

FER = [0.000,0.184,0.329,0.450]
S_ber = [77.85,65.79,57.63,45.34]
t_ber = [11742,13892,15859,17126]

Tprop = [0.000,0.010,0.050,0.100]
S_del = [77.85,75.81,69.03,61.99]
t_del = [11742,12055,13240,20158]

C = [1800,9600,38400,115200]
S_baud = [77.84,77.85,77.82,77.80]
t_baud = [62620,11742,2936,979]

def R(t): 
    return round(FILE_SIZE * 1000 / t)

def make_table(var, S, t, header):
    lines = [header]
    for i in range(len(var)):
        lines.append(f"{var[i]} & {S[i]} & {R(t[i])} \\\\")
    return "\n".join(lines)

table_L    = make_table(L,    S_l,    t_l,    r"$L$ & $S(\%)$ & $R$ \\ \hline")
table_BER  = make_table(FER,  S_ber,  t_ber,  r"FER & $S(\%)$ & $R$ \\ \hline")
table_DEL  = make_table(Tprop,S_del,  t_del,  r"$T_{prop}$ & $S(\%)$ & $R$ \\ \hline")
table_BAUD = make_table(C,    S_baud, t_baud, r"$C$ (bps) & $S(\%)$ & $R$ \\ \hline")

latex = """
\\begin{figure}[h!]
\\centering

\\begin{minipage}{0.23\\linewidth}
\\centering
\\begin{tabular}{c c c}
\\hline
%s
\\hline
\\end{tabular}
\\end{minipage}
\\hfill
\\begin{minipage}{0.23\\linewidth}
\\centering
\\begin{tabular}{c c c}
\\hline
%s
\\hline
\\end{tabular}
\\end{minipage}
\\hfill
\\begin{minipage}{0.23\\linewidth}
\\centering
\\begin{tabular}{c c c}
\\hline
%s
\\hline
\\end{tabular}
\\end{minipage}
\\hfill
\\begin{minipage}{0.23\\linewidth}
\\centering
\\begin{tabular}{c c c}
\\hline
%s
\\hline
\\end{tabular}
\\end{minipage}

\\end{figure}
""" % (table_L, table_BER, table_DEL, table_BAUD)

print(latex)
```
